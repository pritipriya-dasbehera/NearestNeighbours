```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
brute_time = pd.read_csv("brutetime.csv")
grid_time = pd.read_csv("timegrid.csv")
```

```python
grid_time.head()
```

```python
x = brute_time.num_of_particles
y = brute_time.t_brute
X = grid_time.NUM_OF_PARTICLES
Y = grid_time.t_gridcalc
```

```python
plt.scatter(x,y)
plt.scatter(X,Y)
```

```python
brute_time2 = pd.read_csv("brutetime2.csv")
grid_time2 = pd.read_csv("timegrid2.csv")
```

```python
x2 = brute_time2.num_of_particles
y2 = brute_time2.t_brute
X2 = grid_time2.NUM_OF_PARTICLES
Y2 = grid_time2.t_gridcalc
```

```python
plt.scatter(x2,y2)
plt.scatter(X2,Y2)
```

```python
fig, axes = plt.subplots(1,2,figsize=(12, 5),dpi=300)

axes[0].set_xlabel("Number of points")
axes[0].set_ylabel("Time required (in secs)")
axes[0].set_title("Number of Nearest neighbours = 32")
axes[0].scatter(x, y/10**6,label="Brute-Force method")
axes[0].scatter(X, Y/10**6,label="Grid method",marker='x')
axes[0].legend()

axes[1].set_xlabel("Number of points")
axes[1].set_ylabel("Time required (in secs)")
axes[1].set_title("Number of Nearest neighbours = 128")
axes[1].scatter(x2, y2/10**6,label="Brute-Force method")
axes[1].scatter(X2, Y2/10**6,label="Grid method",marker='x')
axes[1].legend()
fig.savefig("fig.jpg")
```
